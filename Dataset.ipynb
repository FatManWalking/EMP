```python
def get_infos(graph):
    edges = number_of_edges(graph)
    nodes = number_of_nodes(graph)
    average_degree = np.mean(nx.degree(graph))
    average_neighbor_degree = np.mean(list(nx.average_neighbor_degree(graph).values()))
    try:
        correlation = nx.degree_pearson_correlation_coefficient(graph)
    except ValueError:
        print("ok")

    return [edges, nodes, average_degree, average_neighbor_degree, correlation]
```

```python
def bias(data, labels):
    labels = np.array(labels)
    labels = np.expand_dims(labels, axis=-1)
    labels[labels == 0] = -1
    E_xk = np.nanmean(data, axis=0)
    E_y = np.nanmean(labels)

    E_xky = np.nanmean(data * labels, axis=0)
    top = E_xky - (E_xk * E_y)
    bottom = np.nanvar(data, axis=0) * np.nanvar(labels)

    R = top / bottom

    return R
```
